```python
#Granting acess to the user
granted=False
def grant():
    global granted
    granted =True
    
#Login to the interface by checking if you are already a user from the database while loggin in and registering
#new user in the database
def login(name,password):
    success=False
    file=open("Login_details.txt","r")
    for i in file:
        a,b=i.split(",")
        b=b.strip()
        if(a==name and b==password):
            success=True
            break
    file.close()
    if(success):
        print("Login successful")
        grant()
    else:
        print("Wrong username or password")

        
#Open the file and register the user
def register(name,password):
    file = open("Login_details.txt","a")
    file.write("\n"+name+","+password)
    file.close()
    grant()
    
#Inputting the data from the user
def access(option):
    global name
    if(option=="login"):
        name=input("Enter your name 'Case sensitive':")
        password=input("Enter the password 'Case sensitive':")
        login(name,password)
    else:
        print("Enter your name and Password to register:\n")
        name=input("Enter your name:  ")
        password=input("Enter the password:  ")
        register(name,password)
#Asking the user if they want to login or register
def begin():
    global option
    print("Welcome to JP Morgan's Resume Screening Interface")
    option=input(" Login or Register (login,register):")
    if(option!="login" and option!="reg"):
        begin()
begin()
access(option)
if(granted):
    print("Welcome to JP Morgan's Resume screening Interface")
    print("User Details:")
    print("Username: ",name)
     
```

```python
import pandas as pd
import os
from tqdm import tqdm
import spacy
from spacy.tokens import DocBin

nlp=spacy.blank("en")

db=DocBin()

for text, annot in tqdm(j):
    doc=nlp.make_doc(text)
    ents=[]
    for start,end ,label in annot["entities"]:
        span= doc.char_span(start,end,label=label,alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
            doc.spans["my_spans"] = list(doc.ents) + [span]
            db.add(doc)
        
os.chdir(r'C:\Users\savita\AppData\Local\Programs\Python\Python39\Scripts\Resume screening')
db.to_disk("./train.spacy")
```
